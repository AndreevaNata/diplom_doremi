```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
steps = 100
x = [x for x in range(1, steps)]
```

```python
y_base = pd.read_csv('/root/diplom_doremi/doremi/logs/loss_baseline.csv').values.tolist()
y_doremi = pd.read_csv('/root/diplom_doremi/doremi/logs/loss_doremi.csv').values.tolist()
```

```python
plt.plot(x, y_base, label='Baseline')
plt.plot(x, y_doremi, label='DoReMi')
plt.xlabel('steps') #Подпись для оси х
plt.ylabel('loss') #Подпись для оси y
plt.title('Обучение модели') #Название
plt.legend()
plt.show()
```
